```python
import torch
import matplotlib.pyplot as plt


def f(x, y):
    return (x - 3)**2 + (y - 4)**2

# 梯度下降算法
def gradient_descent(learning_rate=0.01, num_iterations=100):
   
    x = torch.randn(1, requires_grad=True)
    y = torch.randn(1, requires_grad=True)
    
    xs, ys, fs = [], [], []
    
    for i in range(num_iterations):
     
        loss = f(x, y)
        loss.backward()
        
        with torch.no_grad():
            x -= learning_rate * x.grad
            y -= learning_rate * y.grad
            x.grad.zero_()
            y.grad.zero_()
        
     
        xs.append(x.item())
        ys.append(y.item())
        fs.append(loss.item())
    
    return xs, ys, fs


xs, ys, fs = gradient_descent()

# 绘制
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(xs, label='x')
plt.plot(ys, label='y')
plt.legend()
plt.title('Parameter values during iterations')

plt.subplot(1, 2, 2)
plt.plot(fs)
plt.title('Function value during iterations')

plt.show()
```
